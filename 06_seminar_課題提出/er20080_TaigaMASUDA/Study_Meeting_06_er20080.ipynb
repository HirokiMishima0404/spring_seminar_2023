## 課題3
• シグモイド関数,ソフトマックス関数,tanh関数をまとめて表示してください<br>
– 条件:x = np.arange(-5.0, 5.0, 0.1),plt.ylim(-1.2, 1.2),凡例,軸ラベル,タイトル<br>
• グラフのサイズ,凡例,軸ラベル,タイトルは適切なサイズに変更(できれば)

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def softmax(x):
    u = np.sum(np.exp(x))
    return np.exp(x)/u

def tanh(x):
    y = (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
    return y

x = np.arange(-5.0, 5.0, 0.1)

y_0 = sigmoid(x)
y_1 = softmax(x)
y_2 = tanh(x)

plt.plot(x, y_0, label="sigmoid")
plt.plot(x, y_1, label="softmax")
plt.plot(x, y_2, label="tanh")
plt.ylim(-1.2, 1.2)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("activation_function")
plt.grid()
plt.legend()
plt.show()
plt.savefig("activation_function_fig.svg")
plt.savefig("activation_function_fig.png")
plt.savefig("activation_function_fig.pdf")
```
